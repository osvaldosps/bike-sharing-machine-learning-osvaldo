| **Priority**  |**Evaluation Metrics**     | **Description**   | **Indicator** |
| ---           | ---                       | ---               | ---           |
| 1             | MAE                       | Mean Absolute Error digunakan untuk mengukur sejauh mana perbedaan antara nilai yang diprediksi oleh suatu model atau algoritma dengan nilai yang sebenarnya, tanpa memperhatikan arah perbedaan | Semakin kecil nilai MAE (mendekati 0), semakin bagus model nya |
| 2             | RMSE                      | Root Mean Square Error adalah akar rata-rata dari selisih kuadrat antara nilai yang diprediksi dan nilai yang sebenarnya. | 3             | MAPE                      | Mean Absolute Percentage Error digunakan untuk mengukur persentase kesalahan prediksi rata-rata dari suatu model atau algoritma. Metrik ini merupakan metrik yang paling mudah diinterpretasi dan dijelaskan ke masyarakat umum | Semakin kecil persentase kesalahan, semakin baik performa model dalam memprediksi nilai yang sebenarnya | 

Pada pembuatan model machine learning di project ini, metrik yang diprioritaskan adalah MAE karena MAE menghitung perbedaan absolut antara nilai prediksi dan nilai sebenarnya, tanpa menggunakan operasi kuadrat pada error. Dengan tidak menggunakan operasi kuadrat, nilai error yang besar (seperti yang disebabkan oleh outliers) tidak diperbesar secara signifikan. 

Dalam hal ini, outliers tidak memiliki pengaruh yang terlalu besar pada nilai keseluruhan MAE. MAE memberikan perhatian yang lebih seimbang terhadap semua error, termasuk outliers. Hal ini sangat berpengaruh signifikan apabila ingin memprediksi data-data yang sebelumnya tidak ada pada dataset dan dapat memiliki impact yang signifikan terhadap profit/loss yang diperoleh.

Selain itu, diberikan juga perhatian tehadap Standar Deviasi (std) pada setiap metrik evaluasi untuk mengukur sejauh mana kesalahan prediksi tersebar atau bervariasi dari nilai yang sebenarnya. Standar deviasi memberikan gambaran tentang tingkat dispersi atau penyebaran dari residual kesalahan prediksi, sehingga bisa memberikan informasi tentang seberapa stabil atau konsisten model dalam melakukan prediksi.

## **Cross Validation**

Cross-validation adalah sebuah metode evaluasi model untuk mengukur performa model secara objektif. Tujuannya adalah untuk mengestimasi seberapa baik model yang dibangun akan berkinerja pada data yang belum pernah dilihat sebelumnya.

Dalam cross-validation pada pembuatan model di project ini, dataset awal dibagi menjadi 5 subset yang disebut "fold". Prosesnya melibatkan iterasi di mana setiap fold digunakan sebagai data uji (testing) sementara subset lainnya digunakan sebagai data latih (training). 

Model dilatih dengan menggunakan data latih dan kemudian dievaluasi dengan menggunakan data uji. Proses ini diulangi dengan memvariasikan fold mana yang digunakan sebagai data uji hingga semua fold telah digunakan sebagai data uji.

```python
# looping parameter
models = [linreg, lasso, ridge, knn, tree, voting, stacking, bagging, rf, adaboost, gboost, xgboost] 
scoring_method = ['neg_root_mean_squared_error','neg_mean_absolute_error','neg_mean_absolute_percentage_error']


# result
list_cv_score_mean = []
list_cv_score_std = []
list_scoring_methods = []
list_algo = []
list_resampling_methods = []

# looping
for algo in models:
    
    cv_score_mean = []
    cv_score_std = []

    for method in scoring_method:

        pipe_model = Pipeline([
            ('preprocessing', transformer),
            ('modeling', algo)
        ])

        # cross validation
        cv_score = cross_val_score(
            estimator= pipe_model,  
            X = X_train,            
            y = y_train,            
            cv = 5,                 
            scoring= method,
            error_score='raise'
        )

        cv_score_mean.append(cv_score.mean())
        cv_score_mean.append(cv_score.std())

    list_cv_score_mean.append(cv_score_mean)
    list_scoring_methods.append(method)
    list_algo.append(algo)
```

```python
cv_score_columns = ['mean_rmse','std_rmse','mean_mae','std_mae','mean_mape','std_mape']
model_name = ['linreg', 'lasso', 'ridge', 'knn', 'tree', 'voting', 'stacking', 'bagging', 'randomforest', 'adaboost', 'gboost', 'xgboost'] 

df_cv_score = pd.DataFrame(list_cv_score_mean,columns=cv_score_columns)
df_cv_score.insert(loc=0,value=model_name,column='algo')

df_cv_score.sort_values('mean_rmse',ascending=False)
```

## **Hyperparameter Tuning**

Hyperparameter tuning adalah proses mengoptimalkan nilai-nilai hyperparameter dalam sebuah model machine learning untuk mencapai performa yang lebih baik. 

Hyperparameter adalah parameter yang tidak diajarkan oleh model itu sendiri, tetapi harus ditentukan sebelum pelatihan model dimulai. 

Proses hyperparameter tuning melibatkan eksperimen dengan berbagai nilai hyperparameter yang mungkin. Ada beberapa pendekatan yang digunakan untuk hyperparameter tuning di project ini:

- Grid Search: Menciptakan ruang pencarian (grid) dari beberapa nilai hyperparameter yang mungkin, dan secara sistematis mencoba setiap kombinasi nilai hyperparameter tersebut untuk menemukan kombinasi yang menghasilkan performa terbaik.
- Random Search: Menciptakan ruang pencarian (space) dari beberapa nilai hyperparameter yang mungkin, dan secara acak memilih kombinasi nilai hyperparameter untuk dievaluasi. Pendekatan ini lebih efisien daripada Grid Search ketika ruang pencarian hyperparameter besar.

Untuk mengetahui hyperparameter yang tepat, maka dilakukan perbandingan antara nilai RMSE, MAE, dan MAPE dari masing-masing hasil tuning. Masing-masing model akan dilakukan hyperparameter tuning sebanyak 5 kali, dengan memadukan Grid Search dan Random Search guna mengefesiensikan waktu pencarian.


```python
def tuning_result(result):
    rmse = result.loc[0,'mean_test_neg_root_mean_squared_error']
    mae = result.loc[0,'mean_test_neg_mean_absolute_error']
    mape = result.loc[0,'mean_test_neg_mean_absolute_percentage_error']

    tuning_result = [rmse,mae,mape]

    return tuning_result
```

### **XGBoost**

XGBoost (Extreme Gradient Boosting) adalah sebuah algoritma ensemble learning yang populer dan kuat untuk tugas-tugas regresi dan klasifikasi. XGBoost merupakan implementasi dari metode gradient boosting yang dioptimalkan untuk kinerja dan skalabilitas.

Konsep utama di balik XGBoost adalah menggabungkan beberapa model prediktif yang lemah (weak learner) menjadi model yang lebih kuat (strong learner). XGBoost menggunakan pendekatan boosting, di mana setiap model berikutnya dibangun untuk memperbaiki kesalahan prediksi model sebelumnya. Dalam hal ini, model-model yang dihasilkan disebut "weak learner" atau "base learner".

Berikut adalah langkah-langkah umum dalam cara kerja XGBoost:

1) **Inisialisasi Model**: Model awal dibangun dengan menggunakan weak learner, biasanya dalam bentuk decision tree yang dangkal atau regresi linier sederhana. Weak learner ini akan menjadi basis atau baseline model.
2) **Perhitungan Residuals**: Residuals atau kesalahan prediksi dari model awal dihitung sebagai selisih antara nilai prediksi dan nilai sebenarnya dari data pelatihan.
3) **Membangun Weak Learner Berikutnya**: Weak learner berikutnya dibangun untuk memprediksi residuals dari model sebelumnya. Tujuannya adalah untuk memperbaiki kesalahan prediksi model sebelumnya.
4) **Penyesuaian Model**: Weak learner baru ditambahkan ke model awal, sehingga model menjadi lebih kompleks dan mampu memperbaiki kesalahan prediksi. Pemberian bobot atau penyesuaian terhadap weak learner ini dilakukan dengan memperhatikan kepentingan (importance) setiap fitur dalam model.
5) **Gradient Descent**: XGBoost menggunakan metode gradient descent untuk meminimalkan fungsi kerugian (loss function). Gradient descent digunakan untuk menentukan setiap parameter model dengan memperhitungkan gradien dari fungsi kerugian terhadap parameter tersebut.
6) **Regularisasi**: Untuk menghindari overfitting, XGBoost menerapkan regularisasi dengan memasukkan fungsi penalti L1 dan L2 ke dalam fungsi kerugian. Regularisasi membantu dalam mengendalikan kompleksitas model dan mencegah terjadinya overfitting.
7) **Pembelajaran Berurutan**: Proses langkah 2 hingga 6 diulang untuk membangun weak learner berikutnya dan memperbaiki model secara bertahap. Setiap weak learner berikutnya memfokuskan pada residu dari model sebelumnya, sehingga secara kumulatif mengurangi kesalahan prediksi.
8) **Prediksi**: Setelah model selesai dibangun, prediksi dilakukan dengan menggabungkan hasil prediksi dari semua weak learner. Biasanya, prediksi akhir adalah agregat dari prediksi masing-masing weak learner, seperti melakukan voting atau menghitung rata-rata.

Selama proses ini, XGBoost juga menerapkan beberapa teknik khusus seperti penanganan missing values, skala besar, dan regularisasi untuk meningkatkan performa model dan menghindari overfitting.

Melalui pendekatan ini, XGBoost dapat membangun model yang lebih kompleks dan akurat dengan memanfaatkan kekuatan dari beberapa weak learner dan memperbaiki kesalahan prediksi secara berurutan.

## **Profit Calculation**

### **Business Case Assumption**

- Harga sewa sebuah sepeda per customer baik terdaftar atau tidak = 5 dollar
- Biaya persiapan dan pengiriman sepeda ke titik penyewaan apabila banyak sepeda ditentukan 3 hari sebelumnya = 3 dollar
- Biaya persiapan dan pengiriman sepeda ke titik penyewaan apabila banyak sepeda ditentukan pada hari yang sama = 6 dollar
- Sebelum melakukan machine learning, perusahaan menyediakan sepeda per hari sebanyak rata-rata penyewaan sepeda sebelumnya. Apabila kekurangan, maka perusahaan akan menyediakan sepeda pada hari yang sama
- Dengan menggunakan machine learning, apabila ada prediksi banyak nya sepeda yang disewakan lebih sedikit daripada demand sebenarnya, maka diasumsikan bahwa pelanggan bersedia menunggu kedatangan sepeda ke docking station pada hari yang sama (sehingga tidak terjadi kehilangan sales/pelanggan tidak cancel order)
- Profit dihitung dengan menghitung selisih antara revenue dan operational cost, apabila nilainya negatif berarti mengalami kerugian (Loss)

```python
# harga rental per sepeda
rent_price = 5           

# biaya operating cost (advanced delivery)
normal_processing = 3

# biaya operating cost (immediate delivery)
immediate_processing = 6
```

```python
# median of bike based on historical data (train)
bike_daily = y_train.median()
bike_daily
```

```python
# function untuk menghitung biaya processing

def processing_cost_calc(x):
    if x<0:
        return x * immediate_processing * (-1)
    else:
        return 0
```

```python
# function untuk menghitung profit

def profit_calc(prediction):    
    df = pd.DataFrame()
    df['actual_rent'] = y_test
    df['bike_available'] = prediction
    df['diff'] = df['bike_available'] - df['actual_rent']
    df['normal_processing_cost'] = df['bike_available']*normal_processing
    df['add_processing_cost'] = df['diff'].apply(processing_cost_calc)
    df['revenue'] = df['actual_rent']*rent_price
    df['profit'] = df['revenue'] - (df['normal_processing_cost'] + df['add_processing_cost'])
    return df
```
